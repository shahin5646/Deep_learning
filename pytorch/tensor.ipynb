```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def bark(self):
        print(f"{self.name} is barking........")

    def info(self):
        print(f"{self.name} is a {self.breed}")


german_shepherd = Dog("Rex", "German Shepherd")
german_shepherd.bark()
german_shepherd.info()
```
